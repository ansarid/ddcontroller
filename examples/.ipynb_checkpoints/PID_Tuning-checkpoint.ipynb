# Lab 3: Compass Heading

## Reading the Raw Magnetometer Values

```python
%matplotlib notebook

import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import math
import time
from ddcontroller import DDRobot
```

```python
fig, ax = plt.subplots()

plt.title("Heading")
plt.xlabel("Time (s)")
plt.ylabel("Heading (degrees)")
plt.ylim(-200,200)

times = []
target = []
heading = []

# Create robot object
robot = DDRobot(debug=True, config_path="default.yaml")

def animate(i):
    
    # Set robot's heading to pi/2 with a maximum angular velocity of 1 rad/s
    robot.set_heading(math.pi/2, max_angular_velocity=1)
    
    target.append(robot.target_heading)
    heading.append(robot.get_heading())

    times.append(time.monotonic()-start_time)
    
    plt.scatter(times, target, s=1, color='r')
    plt.scatter(times, heading, s=1, color='b')

start_time = time.monotonic()

times.clear()
target.clear()
heading.clear()

ani = animation.FuncAnimation(fig, animate, interval=1/50)
plt.show()
```
